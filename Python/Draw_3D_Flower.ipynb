# Flower

```python
import numpy as np
from scipy import constants
import matplotlib.pyplot as plt 

n = 800
A = 1.995653
B = 1.27689
C = 8

r = np.linspace(0,1,n)
theta = np.linspace(-2,20*constants.pi,n)
R , THETA =np.meshgrid(r,theta)
petalNum=3.6
x = 1 - (1/2)*((5/4)*(1 - np.remainder(petalNum*THETA, 2*np.pi)/np.pi)**2 - 1/4)**2
phi = (constants.pi/2)*np.exp(-THETA/(C*constants.pi))
y = A*(R**2)*(B*R - 1)**2*np.sin(phi)
R2 = x*(R*np.sin(phi) + y*np.cos(phi));
X=R2*np.sin(THETA);
Y=R2*np.cos(THETA);
Z=x*(R*np.cos(phi)-y*np.sin(phi));
red_map=np.linspace(1,0.25,num=10)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, color='r')
```
