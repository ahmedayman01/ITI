```python
#import Libaries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from matplotlib import gridspec
```

```python
#Read input dataset
col_names = ['sepal_length','sepal_width','petal_length','petal_width', 'class']
df = pd.read_csv("iris.data", header=None)
df.columns = col_names
df
```

```python
#Describe dataset
df.describe()
```

```python
#get info about dataset
df.info()
```

```python
#check for missing data
df.isnull().sum()
```

```python
#drop duplicate data instances if any
df.drop_duplicates().any()
```

```python
#find number of unique classes
df['class'].unique()
```

```python
#Check if dataset is balanced
df.groupby('class').size()
```

# Visualization visual inspection of dataset


```python
#Box Plot for each input variable
pd.plotting.boxplot_frame(df,figsize=(10,10), color='blue', vert=False)
plt.show()
```

```python
sns.boxplot(data = df)
plt.show()
```

```python
#Histogram Plot for each input variable

pd.plotting.hist_frame(df,figsize=(20,20))
plt.show()
```

```python
a = list(df.columns.difference(['class']))
fig, axes = plt.subplots(4, 4, figsize=(20, 22))
for i in a:
    # fig = plt.figure(figsize=(18, 9))
    # spec = gridspec.GridSpec(ncols=3, nrows=2, figure=fig)
    sns.set_style("white")
    # ax = fig.add_subplot(spec[1,0])
    plt.title(i)
    sns.stripplot(x="class", y=i, data=df)
    plt.tight_layout()


plt.show()
```

```python
X=df[df.columns.difference(['class'])]
```

```python
target=pd.factorize(df['class'])[0]
#scatterÂ PlotÂ forÂ twoÂ inputÂ variables
pd.plotting.scatter_matrix(X,figsize=(15,15),marker='o',hist_kwds={'bins':20},s=60,alpha=.8,c=target)
plt.legend()
plt.show()
```

```python
df.sepal_width.plot(kind = 'line', color = 'red', linewidth=1,alpha = 0.5,grid = True,linestyle = ':', figsize = (10,10))
df.sepal_length.plot(kind = 'line', color = 'green', linewidth=1,alpha = 0.5,grid = True,linestyle = '-.', figsize = (10,10))
df.petal_width.plot(kind = 'line', color = 'blue', linewidth=1,alpha = 0.5,grid = True,linestyle = '--', figsize = (10,10))
df.petal_length.plot(kind = 'line', color = 'black', linewidth=1,alpha = 0.5,grid = True,linestyle = '-', figsize = (10,10))
plt.legend()
plt.title('Line Plot')
plt.show()
```

```python
clss = df['class'].value_counts()
plt.title("Iris Sampels Count")
clss.plot(kind='bar', figsize=(10, 6))
```

```python
ax=plt.subplots(1,1,figsize=(10,8))
clss.plot.pie(autopct='%1.1f%%',shadow=True,labels=None,figsize=(10,8))
plt.title("Iris Sampels Ratio")
plt.legend(labels=df['class'], loc='upper left') 
plt.show()
```
