```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.metrics import mean_absolute_error
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_classification
from statsmodels.api import OLS


# Select categorical columns with relatively low cardinality (convenient but arbitrary)
categorical_cols = [cname for cname in df.columns if
                     
                    df[cname].dtype == "object"]

# Select numerical columns
numerical_cols = [cname for cname in df.columns if 
                df[cname].dtype in ['int64', 'float64']]

```

```python
# Apply ordinal encoder to each column with categorical data
ordinal_encoder = OrdinalEncoder()
df[categorical_cols] = ordinal_encoder.fit_transform(df[categorical_cols])
# x_test[categorical_cols] = ordinal_encoder.transform(x_test[categorical_cols])
```

```python
corr_with_y = df.corr()['Apparent Temperature (C)'].drop('Apparent Temperature (C)')
corr_with_y
```

```python
features = ['Precip Type','Temperature (C)','Humidity','Wind Speed (km/h)','Wind Bearing (degrees)','Visibility (km)','Daily Summary']
```

```python
from sklearn.model_selection import train_test_split
y= df['Apparent Temperature (C)']
x= df[features]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
x_train
```

```python
def score(x_test,y_test,model):
    
    pred= model.predict(x_test)
    print('R2:', sklearn.metrics.r2_score(y_test, pred))
    r2ad = 1 - ( 1-model.score(x_test, y_test) ) * ( len(y_test) - 1 ) / ( len(y_test) - x_test.shape[1] - 1 )
    print('AR2:', r2ad)
    print('MAE:', mean_absolute_error(y_test, pred))

    print(OLS(y_test,x_test).fit().summary())
```

# Linear

```python
from sklearn.linear_model import LinearRegression

linear_x_train = x_train[['Temperature (C)']]
linear_x_test = x_test[['Temperature (C)']]

lin_m = make_pipeline(StandardScaler(),LinearRegression())
lin_m.fit(linear_x_train, y_train)

score(linear_x_test,y_test,lin_m)
```

# Multiple

```python
from sklearn.linear_model import LinearRegression

multiple_m = make_pipeline(StandardScaler(),LinearRegression())
multiple_m.fit(x_train, y_train)

score(x_test,y_test,multiple_m)
```

# Polynomial

```python
from sklearn.preprocessing import PolynomialFeatures

poly_regs= PolynomialFeatures(degree= 5)  
x_train_poly= poly_regs.fit_transform(x_train)
x_test_poly= poly_regs.fit_transform(x_test)

# with sklearn
Polynomial_m = make_pipeline(StandardScaler(),LinearRegression())
Polynomial_m.fit(x_train_poly, y_train)


score(x_test_poly,y_test,Polynomial_m)
```

# SVR

```python
from sklearn.svm import LinearSVR


SVR_m = make_pipeline(StandardScaler(),LinearSVR(random_state=0))
SVR_m.fit(x_train, y_train)


score(x_test,y_test,SVR_m)
```
